```python
df['UHR']=df['ZEIT'].dt.hour
```

```python
df.UHR.value_counts()
```

```python
df['Arbeitszeit'] = df.loc[(df['UHR'] < 18) & (df['UHR'] >= 6),'UHR']
```

```python
df.Arbeitszeit = df.Arbeitszeit.fillna(0)
```

```python
df.Arbeitszeit = df.Arbeitszeit.astype(bool)
```

```python
df.Arbeitszeit.value_counts()
```

```python
df.rename(columns={'Arbeitszeit':'IST_ARBEITSZEIT'},inplace=True)
```

```python
sns.lineplot(x=df['UHR'].unique(),y=df['UHR'].value_counts())
sns.set(rc = {'figure.figsize':(20,5)})
plt.xticks(np.arange(0, 23, step=1),rotation=0);
# Verteilung nach Uhr
```

```python
sns.lineplot(x=df['MONAT'].unique(),y=df.groupby('MONAT')['MONAT'].count().values)
sns.set(rc = {'figure.figsize':(20,10)})
plt.xticks(df['MONAT'].unique(),rotation=90);
# Verteilung nach Monaten
```
